```python
# Import library
import tensorflow as tf
from keras.layers import Conv2D, Dense, Flatten, MaxPooling2D
from keras.models import Sequential

# Import Drive
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    directory='/content/Dataset/rockpaperscisors',
    image_size=(256,256),
    batch_size=32
)
```

```python
data_train = dataset.take(round(len(dataset)*0.70))
data_validation = dataset.skip(round(len(dataset)*0.70)).take(round(len(dataset)*0.2))
data_test = dataset.skip(round(len(dataset)*0.9))
```

```python
model_AI = Sequential([
    Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(256,256,3)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')
])

model_AI.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model_AI.fit(data_train, epochs=10, validation_data=data_validation)
```

```python
model_AI_2 = Sequential([
    Conv2D(filters=16, kernel_size=(3,3), activation='relu', input_shape=(256,256,3)),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(filters=8, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')
])

model_AI_2.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history2 = model_AI_2.fit(data_train, epochs=10, validation_data=data_validation)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label="Model 1")
plt.plot(history2.history['accuracy'], label="Model 2")
plt.title("MODEL AI ACCURACY")
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'], label="Model 1")
plt.plot(history2.history['loss'], label="Model 2")
plt.title("MODEL AI LOSS")
plt.legend()
plt.show()
```

```python
plt.plot(history.history['val_accuracy'], label="Model 1")
plt.plot(history2.history['val_accuracy'], label="Model 2")
plt.title("MODEL AI VALIDATION ACCURACY")
plt.legend()
plt.show()
```
